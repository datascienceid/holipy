```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import os
import pandas as pd
import numpy as np
import datetime as dt
```

```python
chromedriver = os.getcwd()+"/chromedriver.exe"
os.environ['webdriver.chrome.driver'] = chromedriver
```

```python
month_num={'Januari':1,'Februari':2,'Maret':3,'April':4,'Mei':5,'Juni':6,'Juli':7,'Agustus':8,'September':9,'Oktober':10,'November':11,'Desember':12}
```

```python
dic_hol={}
for y in [2013+i for i in range(7)]:
    url_l="http://www.liburnasional.com/kalender-"+str(y)+"/"
    browser=webdriver.Chrome(chromedriver)
    browser.get(url_l)
    time.sleep(10)
    l1=browser.find_element_by_xpath("//div[@class='row row-eq-height libnas-content']").text
    dic_hol[y]=l1
    browser.close()
```

```python
k3=pd.DataFrame()
for y in [2013+i for i in range(6)]:
    k=dic_hol[y]
    k2=k.split('  ')
    for ki in k2:
        dh=ki.split('\n')
        if '-' in dh[2]:
            m1=dh[2]
            m=m1.split('-')
            m2=[]
            for im in m:
                m2.append(int(im.strip().split(' ')[0]))
            m2=list(np.arange(int(m2[0]),int(m2[-1])+1,1))
            for im2 in m2:
                dh[2]=str(im2)+' '+m1.split(' ')[-2]+' '+m1.split(' ')[-1]
                k3=k3.append([dh])
        else: 
            k3=k3.append([dh])
```

```python
k3.columns=['HoliDay','Dayname','Date','non']
k3.index=[i for i in range(len(k3))]
del k3['non']
del k3['Dayname']
k3['Date']=k3.Date.apply(lambda m :m.split(' ')[2]+'-'+str(month_num[m.split(' ')[1]])+'-'+m.split(' ')[0])
k3['date_holiday']=k3.Date.apply(lambda x :dt.datetime.strptime(x, '%Y-%m-%d'))
k3['month_holiday']=k3.date_holiday.apply(lambda x: x.month)
k3['year_holiday']=k3.date_holiday.apply(lambda x: x.year)
del k3['Date']
```

```python
to_english_name={'Tahun Baru Masehi':'New Year Masehi',
                 'Maulid Nabi':'Mawlid',
                 'Tahun Baru Imlek':'Chinese New Year',
                 'Hari Raya Nyepi':'Day of Silence',
                 'Jumat Agung':'Good Friday',
                 'Kenaikan Isa Almasih':'Ascension Day',
                 'Hari Raya Waisak': "Buddha's Birthday",
                 'Isra Miraj': "Isra Mi'raj",
                 'Cuti Bersama':"Leave",
                 'Idul Fitri':"Eid Fitr",
                 'Hari Kemerdekaan':"Indonesia's Independence Day",
                 'Idul Adha':"Eid Adha",
                 'Tahun Baru Hijriyah':'Islamic New Year',
                 'Hari Natal':"Christmas Day" ,
                 'Pemilu': "National Election Day" ,
                 'Hari Buruh': "Labor's day",
                 'Pilkada': "Regional Election Day",
                 'Hari Lahir Pancasila': "Pancasila's Birthday"}
```

```python
k3['HoliDay_E']=k3.HoliDay.apply(lambda x : to_english_name[x])
```
